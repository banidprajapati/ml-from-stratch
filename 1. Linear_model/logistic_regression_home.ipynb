```python
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
```

```python
data = datasets.make_blobs(n_samples=100, n_features=3, random_state=21)
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot of two features from make_blobs")
plt.show()
```
