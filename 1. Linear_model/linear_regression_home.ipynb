```python
# Input features and target values
X = np.random.uniform(20,80,100)
y = np.random.uniform(20,80,100)

# Add bias term (intercept) by adding a column of 1s to X
X_b = np.c_[np.ones(X.shape[0]), X] 
```

# Step 3: Initialize Parameters

```python
# Initialize theta (weights) randomly: shape (2, 1) => [theta_0, theta_1]
theta = np.random.randn(2, 1)
print(theta)
```
